```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_paths = ["C:/Users/FreeComp/Desktop/open/photo_1.jpg", "C:/Users/FreeComp/Desktop/open/photo_2.jpg", "C:/Users/FreeComp/Desktop/open/photo_3.jpg","C:/Users/FreeComp/Desktop/open/photo_4.jpg", "C:/Users/FreeComp/Desktop/open/photo_5.jpg", "C:/Users/FreeComp/Desktop/open/photo_6.jpg", "C:/Users/FreeComp/Desktop/open/photo_7.jpg", "C:/Users/FreeComp/Desktop/open/photo_8.jpg"]
images = [cv2.imread(path, cv2.IMREAD_GRAYSCALE) for path in image_paths]

binary_images = [cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)[1] for image in images]

otsu_images = [cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1] for image in images]

MEAN_adaptive_images=[cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2) for image in images]

GAUSSIAN_adaptive_images = [cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2) for image in images]

plt.figure(figsize=(16, 8))

for i in range(8):
    plt.subplot(5, 8, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(f'Original {i+1} \n')
    
    plt.subplot(5, 8, i + 9)
    plt.imshow(binary_images[i], cmap='gray')
    plt.title(f'Thresholded {i+1} \n')

    plt.subplot(5, 8, i + 17)
    plt.imshow(otsu_images[i], cmap='gray')
    plt.title(f'OTSU {i+1} \n')
    
    plt.subplot(5, 8, i + 25)
    plt.imshow(MEAN_adaptive_images[i], cmap='gray')
    plt.title(f'MEAN {i+1} \n')
    
    plt.subplot(5, 8, i + 33)
    plt.imshow(GAUSSIAN_adaptive_images[i], cmap='gray')
    plt.title(f'GAUSSIAN {i+1} \n')
    
plt.show()
```
